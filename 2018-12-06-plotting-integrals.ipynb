### implementation for intervals on $\mathbf{R}$

- test for $a \le b$

```python
def based_integral(a,b):
    if b >= a:
        return plot(f, (x,0,4)) +\
            plot(f, (x,a,b), fill=True) +\
            plot(lambda x: numerical_integral(f,a,x)[0],(x,0,4), color = "red") +\
            plot(lambda x: numerical_integral(f,a,x)[0],(x,a,b), color = "red", thickness=5) +\
            point((b,numerical_integral(f,a,b)[0]), size=100, color="black")
    else:
        return plot(f, (x,0,4)) +\
            plot(f, (x,b,a), fill=True) +\
            plot(lambda x: numerical_integral(f,a,x)[0],(x,0,4), color = "red") +\
            plot(lambda x: numerical_integral(f,a,x)[0],(x,b,a), color = "red", thickness=5)+\
            point((b,numerical_integral(f,a,b)[0]), size=100, color="black")
```

```python
based_integral(2,1)
```

```python
based_integral(2,3)
```

### two free parameters for integration on a closed interval $[x_0,x_1]$

- integration defined on $[x_0, x_1]$ for $x_0 \le x_1$
    - $x_0$ (base point)
    - $x_1$ (end point)
    
### to integrate $\int_{x_0}^{x_1} f(x)\,dx$, one needs to *order* the bounds of integration

### physical interpretation: conservation of energy 


$$\int_{x_0}^{x_1} f(x)\,dx + \int_{x_1}^{x_0} f(x)\,dx = 0$$

```python
p=[]
for i in range(3):
    p.append(based_integral(1,2+i))
sum(p) 
```

```python
p=[]
for i in range(3):
    p.append(based_integral(4,3-i))
sum(p) 
```

```python
p=[]
for i in range(3):
    p.append(based_integral(4,3-i))
    p.append(based_integral(1,2+i))
sum(p) 
```

### definite integrals: *signed values* of *paths*

- $[a,b]$ is the *path* in $\mathbf{R}$

    - basepoint $a$
    - endpoint $b$
    - always $a < b$ (reorder if you must)
    
- $\int_a^b f(x)\,dx$ is the *signed valued* of the path $[a,b]$ equipped with the weight $f(x)$

### second fundamental theorem of calculus

**Given:** Let $f$ be a real-valued function on a closed interval $[a , b]$  $F$ an antiderivative of $f$ on $[a,b]$ such that $$F'(x)=f(x).$$

**To prove:** If $f$ is *integrable* on $[a,b]$, then ... ***what?***

$$F(b) - F(a) = \int_b^a f(x) \, dx$$

```python
based_integral(.5,1) + based_integral(.5,2.5)
```

```python
based_integral(3.5,1) + based_integral(3.5,2.5)
```
