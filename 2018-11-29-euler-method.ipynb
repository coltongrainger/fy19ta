### Problem 4: Newton's Law of Cooling

Supplemental plots to <https://math.colorado.edu/math1300/projects/1300Project13FTC.pdf>.

```python
def f(y, t):
    return -25*e^(-2*t)
```

```python
def euler (f, t0, y0, h, tmax):
    t, y = t0, y0
    td, yd = [t0], [y0]
    while t < tmax:
        y = y + h*f(y, t)
        yd.append(y)
        t = t + h
        td.append(t)
    return(zip(td,yd))
```

```python
list_plot(euler(f,0,60,1,3), plotjoined=True)
```

Now we plot our approximation against the slope field.

```python
y, t = var('y t')
h = 1
apprx = list_plot(euler(f,0,60,h,3), plotjoined=True)
slope = plot_slope_field(-25*e^(-2*t), (t,0,3), (y,0,60))
plot(apprx + slope)
```

How much does our approximation improve when when decrease the step size by half?

```python
h = 0.5
apprx = list_plot(euler(f,0,60,h,3), plotjoined=True, color="purple")
slope = plot_slope_field(-25*e^(-2*t), (t,0,3), (y,0,60))
plot(apprx + slope)
```
